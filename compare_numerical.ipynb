```python
import matplotlib.pyplot as plt
import re

# 获取所有类名
all_class_names = [
    'background', 'building', 'fence', 'person', 'pole', 
    'road', 'sidewalk', 'vegetation', 'car', 'wall', 'traffic sign'
]

def parse_log_file(file_path):
    results = {}
    results['mIoU'] = {}
    for class_name in all_class_names:
        results[class_name] = {}

    with open(file_path, 'r') as file:
        lines = file.readlines()

    for i, line in enumerate(lines):
        # 提取 epoch 和 mean_IoU
        # epoch_match = re.search(r'Current epoch:(\d+)\s+mIoU:\s+([\d.]+)', line)
        # Current epoch:20 mIoU: 56.44 Best mIoU: 56.44
        epoch_match = re.search(r'Current epoch:(\d+)\s+mIoU:\s+([\d.]+)\s+Best mIoU:\s+([\d.]+)', line)
        if epoch_match:
            epoch = int(epoch_match.group(1))
            mean_iou = float(epoch_match.group(2))
            best_miou = float(epoch_match.group(3))
            results['mIoU'][epoch] = mean_iou
            results['Best_mIoU'] = best_miou


        # 提取每一类的结果
        class_match = re.match(r'(\d+)\s+([\w\s]+)\s+([\d.]+)\s+([\d.]+)', line)
        if class_match:
            class_id = int(class_match.group(1))
            class_name = class_match.group(2).strip()  # 去除多余的空格
            iou = float(class_match.group(3))
            acc = float(class_match.group(4))
            epoch_line_index = i + (12 - class_id)

            if epoch_line_index < len(lines):
                epoch_line = lines[epoch_line_index]
                epoch_match = re.search(r'== Epoch:\s+(\d+)\s+mean_IoU\s+([\d.]+)', epoch_line)
                if epoch_match:
                    epoch = int(epoch_match.group(1))
                    results[class_name][epoch] = {'iou': iou, 'acc': acc}

    return results

# Define the file path
results_dicts = {}
import os
file_dir = 'output/DSEC_CMNeXt-B2_i/'
# file_dir下所有以rgb_day_11_2024开头的文件
# file_paths = sorted(os.path.join(file_dir, file) for file in os.listdir(file_dir) if file.startswith('rgb_day_11_2024'))
file_paths = [
    # # rgb_t + event_t ➡️ seg_{t+1} train from strach 69.18
    # 'output/DSEC_CMNeXt-B2_ie/rgbe_day_11_20241005_154538_train.log',
    # # rgb_{t} ➡️ seg_{t+1} train from strach 66.72
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241005_181400_train.log',
    # # rgb_t ➡️ seg_{t+1} FRMA num_flow=1 lr/10 frozen segformer 
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241014_184051_train.log',
    # # rgb_t ➡️ seg_{t+1} FRMA num_flow=4 lr/10 frozen segformer 
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241014_230458_train.log',
    # # 1️⃣ rgb_t + flow_t ➡️ seg_{t+1} lr/10 train softsplat 63.18
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241015_024633_train.log',
    # # 2️⃣ rgb_t + flow_t ➡️ seg_{t+1} lr/10 train softsplat          , with feature_after residual feature_before 63.59
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241015_024827_train.log',
    # # 3️⃣ rgb_t + flow_t ➡️ seg_{t+1} lr/10 train softsplat & decoder 64.64
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241015_025023_train.log',
    # # 4️⃣ rgb_t + flow_t ➡️ seg_{t+1} lr/10 train softsplat & decoder, with feature_after residual feature_before 64.78
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241015_025203_train.log',
    # # 5️⃣ rgb_t + flow_t ➡️ seg_{t+1} lr  softsplat & decoder, with feature_after residual feature_before 67.92
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241015_145340_train.log',
    # # 6️⃣ rgb_t + flow_t ➡️ seg_{t+1} lr  softsplat & decoder, with feature_after residual feature_before no feature loss 67.88
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241015_145435_train.log',
    # # # 1_1: only rgb_f_warp before refine net 021351 / 024125 68.03
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241016_024125_train.log', 
    # # 1_2: only rgb_f_warp before refine net , using netSoftmetric 033022 70.58
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241016_033022_train.log',
    # # 2: cancate rgb_f & rgb_f_warp before refine net 024209 67.85
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241016_024209_train.log', 
    # # 3: cancate event_f & rgb_f & meticone before softsplat, using netSoftmetric 031944 
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241016_201106_train.log',
    # # 4_1: cancate event_f & rgb_f & rgb_f_warp before refine net 031314 70.03
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241016_031314_train.log',
    # # 4_2: cancate event_f & rgb_f & rgb_f_warp before refine net, using netSoftmetric 031944 70.62
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241016_031944_train.log',
    # # only F before softsplat, F' into refinenet
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241017_043904_train.log',
    # # only F before softsplat, F'+F into refinenet
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241017_043911_train.log',
    # # only F before softsplat, F'+E into refinenet
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241017_043919_train.log',
    # # only F before softsplat, F'+E+F into refinenet
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241017_043925_train.log',
    # # [F, w] [F, w]+[F', w']
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241017_171212_train.log',
    # # [F, w] [F', w'] with photometric loss
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241017_171318_train.log',
    # # [F, w] [F', w'] with photometric loss; random forward backward
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241017_181536_train.log',
    # # unfrozen backbone
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241017_235818_train.log',
    # # forward flow then backward img
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241018_012800_train.log',
    # # use 7 para for metric
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241018_144152_train.log',
    # # 6e-5
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241018_153205_train.log',
    # # 8e-5
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241018_153538_train.log',
    # # 8e-5 lr for forward splat 70.8
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241018_215427_train.log',
    # # 1e-4 lr for forward splat 70.86
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241019_011739_train.log',
    # # 1e-4 lr for forward splat + more conv 71.64
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241019_011919_train.log',
    # # 1e-4 lr for forward splat + more conv + sigmoid
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241019_020450_train.log',
    # # 1e-4 lr for forward splat + sigmoid + rgb
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241019_024034_train.log',
    # # 1e-4 lr for forward splat + more more conv bs 6*2 400 epoch 71.74
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241019_143233_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_150052_train.log',
    # #  1e-4 lr for forward splat + more more conv  bs 8*2 400 epoch 71.66
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241019_185028_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_144920_train.log',
    # 1e-4 lr for forward splat + more more conv 200 epoch
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241019_080516_train.log',
    # # 1e-4 lr for forward splat + more more more conv 400 epoch
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241019_080612_train.log',
    # # 1.5e-4 lr for forward splat + more more conv 400 epoch power 0.9 warmup 5 bs 16*2
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_161301_train.log',
    # # 1e-4 lr more-more conv, epoch 200, bs 6*2; power 0.85 warm 10 
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_073436_train.log',
    # # 1e-4 lr more-more conv, epoch 200, bs 6*2; power 0.85 warm 5
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_073452_train.log',
    # # 1e-4 lr more-more conv, epoch 200, bs 6*2; power 0.90 warm 10 cos scheduler
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_073904_train.log',
    # # warp + SENet; event c flow for metric 71.55
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_191540_train.log',
    # # warp + SENet; rgb c flow for metric
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_192239_train.log',
    # # warp + SENet; rgb c event c flow for metric
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_113623_train.log',
    # # warp; event for metric 71.50
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_203246_train.log',
    # # warp; flow for metric
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_123412_train.log',
    # # more-more conv for netOut in metric net
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_142450_train.log',
    # # # more channel for netOut in metric net
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_143234_train.log',
    # # # add event feature extractor
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241020_231946_train.log'
    # # bin 4 71.81
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_010243_train.log',
    # # bin 2
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_010431_train.log',
    # # bin 10
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_010614_train.log',
    # # bin 1
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_014825_train.log',
    # # bin 4 with learnable metric scale factor 
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_033151_train.log',
    # # bin 4 with learnable metric after netOut
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_035035_train.log',
    # # with more data aug 71.86
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_044246_train.log',
    # # with residual on before and after 71.68
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_050818_train.log',
    # # with pac
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_051302_train.log',
    # # basic attention se in netSoftmetric
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_152757_train.log',
    # # basic attention cbam in netSoftmetric 71.89
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_152931_train.log',
    # # cross scale
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_165703_train.log',
    # # LeakyReLu
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_182227_train.log',
    # # ELU
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_182454_train.log',
    # # Mish
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_102822_train.log',
    # # warp on x[0]
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241111_052703_train.log',
    # # warp on y
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_103713_train.log',
    # # 0.4 percent zero flow
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_222834_train.log',
    # # correct flow resize scale
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241021_225119_train.log',
    # # correct no flow event norm 72.69 ! 
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241022_172623_train.log',
    # correct no flow norm 72.87 72.93 pesudo gt flow SOTA
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241022_174808_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241026_164158_train.log',  # 72.93
    # # kernel size on refinenet is 5
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241026_184116_train.log',
    # # change refine conv net to multi attention
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241025_213538_train.log',
    # # event atten & img atten
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241025_213830_train.log',
    # # img atten & event atten
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241025_141433_train.log',
    # # img atten
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241025_223849_train.log',
    # # self attention
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241025_235356_train.log',
    # # self attention head 8 
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241025_171629_train.log',
    # # head 8 with attention between feature and metric
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241025_184709_train.log',
    # # self attention head 16
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241026_023038_train.log',
    # # head 16 with attention between feature and metric 4090g 72.17
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241025_184709_train.log',
    # # head 16 with attention between feature and metric num_multi_attn=4 72.1
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241025_190845_train.log',
    # # head 16 with attention between feature and raw metric num_multi_attn=4 72.21
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241026_032833_train.log',
    # # head 16 num+multi_atten 1 rao metric lr 6e-5
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241026_132926_train.log',
    # # dilation 125 125
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241026_135204_train.log',
    # # warp on y 71.63
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241022_101026_train.log',
    # # unet predict flow from img and event
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241024_032207_train.log',
    # # FRMA predict flow from img and event FRMA 4 flow 2
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241025_175437_train.log',
    # # FRMA predict flow from img and event FRMA 2 flow 8 71.48
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241025_105237_train.log',
    # # eraft predict flow using pretrained model SOTA
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241028_020453_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241027_190252_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_002738_train.log',
    # 塌了的
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241028_220121_train.log',
    # # frozen flow net 不如不frozen
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241028_204254_train.log',
    # zero flow 71.30
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241028_151612_train.log',
    # bflow predict flow using pretrained model 塌了
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241028_171437_train.log',
    # self attention bs = 4
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241028_230331_train.log',
    # cross attention on backbone
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241028_231657_train.log',
    # train from scratch t0-t0 73.99
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_035200_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_133947_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_162058_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_163440_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_203154_train.log',  # clip norm 73.50   # NOTE
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_000716_train.log',  # clip norm test new machine
    # train from scratch t0-t1 67.67
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_035802_train.log',
    # train from scratch t0-t2 62.42
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_035441_train.log',
    # eraft flow 100ms event_20
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_172615_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_175722_train.log',
    # 'output=/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_184337_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_143346_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_155446_train.log',  # n_channel=10 in eraft 70.38
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_120727_train.log',  # n_channel=2 in eraft but clip norm maxnorm 20 71.66
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241031_000843_train.log',  # n_channel=2 in eraft but clip norm maxnorm 15 没啥用
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241101_230856_train.log',  # iterative with warp img bs=9 due to gou limit
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241101_233014_train.log',  # iterative with warp img bs=9 due to gou limit
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241101_234739_train.log',  # bs=9 due to gou limit
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241102_003002_train.log',  # bflow test2 img bs=9 due to gou limit
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241102_013146_train.log',  # bflow test2 img bs=9 due to gou limit 72.21 # NOTE
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241102_130737_train.log',  # eraft test2 img bs=9 due to gou limit
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241102_134211_train.log',  # bflow with attention based fusion bs=9 due to gou limit
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241102_135940_train.log'
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241102_172240_train.log',    # bs 12 50ms bflow 0.5 1.0 72.99
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241102_173249_train.log',    # bs 12 100ms bflow 0.5 1.0 71.77
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241102_225109_train.log',    # bs 12 100ms eraft 0.5 1.0 72.99
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241102_232352_train.log',    # bs 12 100ms eraft 0.5 1.0 debug
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241102_233922_train.log',    # bs 12 100ms eraft 1.0
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_013001_train.log',  # new machine test
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_013901_train.log',  # new machine test
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_020125_train.log',  # new machine test
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_020919_train.log',  # new machine test
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_022849_train.log',  # new machine test
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_023624_train.log',  # new machine test bin 40 eraft 71.20
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_031413_train.log',  # new machine test bin 40 bflow
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_140451_train.log',  # new machine test bin 40 bflow
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_135712_train.log',  # new machine test bin 40 bflow total event_voxel for metric nan了
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_034012_train.log',  # new machine test bin 40 eraft with attention fusion 71.59
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_132334_train.log',  # new machine test bin 20 eraft with attention fusion
    # zero flow 100ms event_20
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_160704_train.log',  # 67.88
    # eraft flow 100ms event_40 塌了
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_194519_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241031_023135_train.log',  # iterative bs=9
    # eraft flow 50ms event_20 72.67
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_113014_train.log',  # 72.67
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_181251_train.log',  # clip norm 73.12 # NOTE
    # raft flow 100ms event_20 71.30
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_180305_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241029_183957_train.log',    # 71.30
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_212702_train.log',  # clip norm 71.39
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241030_224508_train.log',  # clip norm & iterative 2 没啥用
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241031_010021_train.log',  # clip norm & iterative 2 debug 没啥用
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_001957_train.log',  # new machine test
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_003742_train.log',  # new machine test
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_010340_train.log',  # new machine test
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241103_012122_train.log',  # new machine test

    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_035304_train.log',  # new machine test# -50-0 0-50 eraft 73.3
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_035414_train.log',  # new machine test# -100-0 0-100 eraft 71.95
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_143543_train.log',  # new machine test# -50-0 0-50 50-100 eraft
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_143607_train.log',  # new machine test# -50-0 0-50 50-100 eraft with mid supervised
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_174219_train.log',  # new machine test# -50-0 0-50 50-100 eraft with flow1+flow2
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241109_165951_train.log',  # new machine test# -50-0 0-50 50-100 eraft with [event2, img2] in cnet
    # bflow iterative 10ms event_20 50ms
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241111_183303_train.log',  # 72.95
    # memory
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241112_022816_train.log',  # 71.97 memory after splat
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20241112_033738_train.log',  # 71.35 memory before refinenet
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250301_235828_train.log',  # event as flow for 100ms
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250302_034145_train.log',  # eraft_large with pretrained; event down 4 times
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250302_154755_train.log',  # eraft_large with pretrained; event down 2 times for flow and 4 times for splat 71.61
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250303_003028_train.log',  # add cls_weights based on the last one 71.85
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250303_101021_train.log',  # event normalization from /255 to /max based on the last one 71.93
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250303_164750_train.log',  # update cls_weights and event normalization based on the last one (calculated before data aug)
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250303_175014_train.log',  # debug flow rescale in frame_synthesis
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250303_201641_train.log',  # residual skip and no event norm
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250303_202223_train.log',  # attention skip and no event norm 结果证明不加norm效果比较好
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250304_053615_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250304_160328_train.log',
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250304_193200_train.log', # more-more-conv-sep bs 4 lr 0.0001
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250304_193556_train.log',  # conv-relu-conv-sep bs 4 lr 0.0001
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250304_203033_train.log',  # conv-relu-conv-sep with He init bs 4 lr 0.0001 71.36
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250304_151810_train.log',  # debug event norm; amp false; bs 4 lr 0.0001
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250304_161942_train.log',  # debug event norm; amp false; bs 8 lr 0.00014 clip norm 1

    # ###### model size aba ######
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_014025_train.log', # half size bs4: raft small & conv-relu-conv-sep & clip bs 12... new training setting 70.28

    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250304_225221_train.log', # half size bs4: raft small & conv-relu-conv-sep & clip 1 70.50
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_061143_train.log', # half size bs4: raft small & conv-relu-conv & clip 1 hku16_2_in_process 71.3
    # # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250304_232049_train.log', # half size bs4: raft small & more-more-conv & clip 1 71.27

    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250304_224650_train.log', # half size bs4: raft large & conv-relu-conv-sep & clip 1 71.41
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_061519_train.log', # half size bs4: raft large & conv-relu-conv & clip 1 hku16_2_in_process 72.08
    # # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_061852_train.log', # half size bs4: raft large & more-more-conv & clip 1 hku16_2_in_process 72.19

    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_052847_train.log', # full size bs4: raft small & conv-relu-conv-sep & clip 1 71.23
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_052811_train.log', # full size bs4: raft small & conv-relu-conv & clip 1 72.28
    # # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_060917_train.log', # full size bs4: raft small & more-more-conv & clip 1 hku16_1_in_process 72.14 (232 72.21)

    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_062143_train.log', # full size bs4: raft large & conv-relu-conv-sep & clip 1 hku16_0_in_process 71.9
    # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_060550_train.log', # full size bs4: raft large & conv-relu-conv & clip 1 cvmi518_in_process 72.64
    'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_141512_train.log',  # full size bs4: raft large & conv-relu-conv & clip 1 cvmi518_0_in_process aba of pad32 and 74.01 pretrained 72.72
    # # 'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_052847_train.log', # full size bs4: raft large & more-more-conv & clip 1 cvmi518_in_process 72.56

    ###### Score Net Ablation ######
    'output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_152359_train.log', # IEF score hku16_1_in_process 7252
    '/home/xy/work/adap_v/segment_anytime/output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_152640_train.log', # IE score hku16_2_in_process 72.51
    '/home/xy/work/adap_v/segment_anytime/output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_152815_train.log', # IF score hku16_4_in_process 72.39 72.61 for 186
    '/home/xy/work/adap_v/segment_anytime/output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_152939_train.log', # F score hku16_5_in_process 72.31
    '/home/xy/work/adap_v/segment_anytime/output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_153742_train.log', # E score hku16_6_in_process 72.03
    '/home/xy/work/adap_v/segment_anytime/output/DSEC_CMNeXt-B2_i/rgb_day_11_20250305_153949_train.log', # I score hku16_7_in_process 72.45

]
for file_path in file_paths:
    parts = file_path.split('/')
    filename = parts[-1]
    # scene = filename.split('_')[:2]
    # scene = '_'.join(scene)
    # print(scene)

    # Initialize an empty dictionary to store the results
    results_dict = parse_log_file(file_path)
    results_dicts[filename] = results_dict
```

```python
def compute_iou(hist):
    ious = hist.diag() / (hist.sum(0) + hist.sum(1) - hist.diag())
    ious[ious.isnan()]=0.
    miou = ious.mean().item()
    # miou = ious[~ious.isnan()].mean().item()
    ious *= 100
    miou *= 100
    return ious, miou
```
